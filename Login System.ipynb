```python
aas# Predefined user accounts (You can modify these)
users = [
    {"user_id": "sarmad", "password": "sarmad123"},
    {"user_id": "asjad", "password": "asjad123"},
    {"user_id": "anzer", "password": "anzer123"}
]

def login_system():
    print("Welcome to the Login System!\n")

    # Allow up to 3 attempts
    attempts = 0
    max_attempts = 3

    while attempts < max_attempts:
        user_id = input("Enter your User ID: ").strip()
        password = input("Enter your Password: ").strip()

        # Search for matching user using a for loop
        user_found = False
        for user in users:
            if user["user_id"] == user_id and user["password"] == password:
                user_found = True
                break

        if user_found:
            print(f"\nWelcome, {user_id}! You have successfully logged in.")
            break  # Exit the loop after successful login
        else:
            attempts += 1
            print(f"Invalid credentials! {max_attempts - attempts} attempt(s) left.\n")

    if attempts == max_attempts:
        print("Maximum attempts reached! Please try again later.")

# Run the login system
login_system()
```
